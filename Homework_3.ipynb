### **1**

`[]`: This method allows for you to access columns or slice rows. I would like to note that you cannot return rows just as asked (1 at a time) without slicing (aka, `[1:2]` to return row the second row (index 1) vs just `[1]`) and you cannot slice columns using this method. You can use compound logics and compare operations to get booleans with this method.


`.iloc[]`: This method of indexing allows you to use the index of a column or row starting from 0 to return rows or columns. This follows the same operating logic as most programming and is set up in a `.iloc[rows,colums]` format so if you want just the column you will have to specify that you want all rows in the third column like the following `.iloc[:,2]`. Use a list to indicate multipule rows or columns. This method does not work with boolean arrays. It is however capable of slicing both rows and columns unlike `[]`.

`.loc[]`: This is a method of using labels to return rows or columns. This method looks for a spesific word or label to indicate a column or row and returns the entire column or row. Confusingly, it includes the last number when doing things like `.loc[1:2]` which would return 2 rows instead of the 1 you would expect. The method looks for something like `.loc["Color"]` because it looks at the labels or column/row names instead of the indexes. This does work with booleans and is capable of sclicing rows and columns just like `.iloc[]`.
